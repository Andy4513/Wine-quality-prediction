## Exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the 3 csv with red wine, white wine and the combined one into dataframes

df_red = pd.read_csv('winequality-red.csv', delimiter=';')
df_white = pd.read_csv('winequality-white.csv', delimiter=';')
df_combined = pd.read_csv('winequality.csv', index_col=0)

df_red.head()
```

```python
df_white.head()
```

```python
df_combined
```

# RED WINE

```python
sns.countplot(x=df_red['quality'])
plt.title('Red wine quality distribution')
plt.show()
```

```python
# Create histograms for each attribute to see the distribution

column_names = df_red.columns

for column in column_names:
    sns.histplot(data = df_red, x = column)
    plt.show()
```

```python
sns.pairplot(df_red, height = 1.5)
plt.show()
```

# White wine

```python
sns.countplot(x=df_white['quality'])
plt.title('White wine quality distribution')
plt.show()
```

```python
sns.pairplot(df_white, height = 1.5)
plt.show()
```

# Combined wines

```python
sns.pairplot(
    df_combined,
    x_vars = df_combined.columns[:-1],
    y_vars = df_combined.columns[:-1],
    height = 1.5
)
plt.show()
```

# Red and white comparison

```python
# Create boxplot comparasion between red and white wine attributes

for column in column_names:
    sns.boxplot(x = df_combined['isRed'], y = df_combined[column])
    plt.show()
```
